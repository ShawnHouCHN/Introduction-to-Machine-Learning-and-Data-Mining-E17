```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
df['goal_difference'] = df['home_team_goal'] - df['away_team_goal']
df = df.drop(labels = ['stage', 'home_team_goal', 'away_team_goal'], axis=1)
df = df.drop(labels = ['BWH','BWD','BWA','IWH','IWD','IWA','LBH','LBD','LBA','WHH','WHD','WHA','VCH','VCD','VCA'], axis=1)
df.head()
```

```python
for columnName in df:
    column = df[columnName]
    limits = column.quantile([0.25, 0.5, 0.75])
    if columnName == 'possession_home':
        df['possession_home_over50'] = df['possession_home'] > 50
    elif columnName == 'goal_difference':
        df['home_win'] = df['goal_difference'] > 0
        df['draw'] = df['goal_difference'] == 0
        df['away_win'] = df['goal_difference'] < 0
    elif columnName.endswith('_isOrganised'):
        df[columnName + '_true'] = column == 1
        df[columnName + '_false']  = column == 0
    else:
        df[columnName + '_verylow'] = column <= limits[0.25]
        df[columnName + '_low']  = (limits[0.25] < column) & (column <= limits[0.5])
        df[columnName + '_high'] = (limits[0.5] < column) & (column <= limits[0.75])
        df[columnName + '_veryhigh'] = column > limits[0.75]
    df = df.drop(columnName, axis=1)
df = df.astype(int)
df.head()
```

```python
df.to_csv('mining.csv', index=False)
```
